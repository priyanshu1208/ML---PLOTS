```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


var =[1,2,3,4,5,6,7]
var_1 = [2,3,4,5,1,6,7]

df = pd.DataFrame({
    "var":var,
    "var_1":var_1
})
df
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
sns.lineplot(x="var",y="var_1",data = df) 
plt.show()
```
