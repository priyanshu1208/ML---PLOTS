```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline 
```

```python
df = pd.read_csv(r"C:\Users\dell\Desktop\titanic.csv")
df.head()
```

```python
sns.barplot(x='Survived',y = 'Pclass',data = df)
plt.show()
```

```python
 sns.barplot(x='Survived',y = 'Pclass',data = df,hue = 'Sex')
plt.show()
```

```python
order_1 = [1,0]
sns.barplot(x='Survived',y = 'Pclass',data = df, order = order_1)
plt.show()
```
